# Preprocess Covid-19 data

The covid data do not need much preprocessing. However, a bit feature engineering is involved. The data we use is from https://github.com/nychealth/coronavirus-data/blob/master/trends/data-by-day.csv and https://github.com/nychealth/coronavirus-data/blob/master/trends/cases-by-day.csv. 

Here, we are only interested in:
- CASE_COUNT: Count of confirmed cases citywide
- BX_CASE_COUNT: Count of confirmed cases in the Bronx
- BK_CASE_COUNT: Count of confirmed cases in Brooklyn
- MN_CASE_COUNT: Count of confirmed cases in Manhattan
- QN_CASE_COUNT: Count of confirmed cases in Queens
- SI_CASE_COUNT: Count of confirmed cases in Staten Island
- DEATH_COUNT: Count of confirmed deaths citywide
- HOSPITALIZED_COUNT: Count of hospitalized cases citywide

```python
# Imports
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import pandas as pd
import seaborn as sns
import glob
import os
import time
%matplotlib inline
from pandas.api.types import CategoricalDtype
import warnings
warnings.filterwarnings("ignore")
```

## Step 1: Import data & feature selection

```python
# import raw data 
cols_of_interest = ["date_of_interest", "CASE_COUNT", "DEATH_COUNT", "HOSPITALIZED_COUNT", \
                    "BX_CASE_COUNT", "BK_CASE_COUNT", "MN_CASE_COUNT", \
                    "QN_CASE_COUNT", "SI_CASE_COUNT"]
covid_raw = pd.read_csv("../raw_data/data-by-day.csv")
covid_raw = covid_raw[["date_of_interest", "CASE_COUNT", "DEATH_COUNT", "HOSPITALIZED_COUNT"]]
covid_raw
```

```python
# import raw data 
covid_raw1 = pd.read_csv("../raw_data/cases-by-day.csv")
covid_raw1 = covid_raw1[["date_of_interest", "BX_CASE_COUNT", "BK_CASE_COUNT", \
                       "MN_CASE_COUNT","QN_CASE_COUNT", "SI_CASE_COUNT"]]
covid_raw1
```

```python
covid_df = pd.merge(covid_raw, covid_raw1, \
                    left_on="date_of_interest", right_on="date_of_interest")
covid_df
```

```python
covid_df.describe()
```

```python
covid_20 = covid_df.iloc[:307] # inspect the csv to find all 2020 data 
covid_21 = covid_df.iloc[308:]
```

## Step 2: Save on desk

## Step 3: Visualize

```python
# daily cases
plt.plot(covid_df.date_of_interest, covid_df.CASE_COUNT, label = 'covid cases')
plt.xlabel('date')
plt.ylabel('case number')
plt.title('Daily confirmed cases in NYC from 2020 to 2021')
plt.rcParams["figure.figsize"] = (30,20)

ax = plt.gca()
formatter = mdates.DateFormatter('%b')
ax.xaxis.set_major_formatter(formatter)
locator = mdates.MonthLocator()
ax.xaxis.set_major_locator(locator)

plt.savefig('../plots/daily_case_number_trend.png')
plt.show()
```
