# Preprocess MTA data

Metropolitan Transportation Authority (MTA) updates systemwide ridership and traffic estimates for subways, buses, Long Island Rail Road, Metro-North Railroad, Access-A-Ride, and Bridges and Tunnels. You can see changes over the past seven days, as well as get a sense of how ridership and traffic differs versus before the pandemic. Subway, bus, Access-A-Ride, and Bridges and Tunnels figures are compared with the equivalent day between 3/1/19 and 2/29/20. LIRR and Metro-North figures are compared with 2019 averages. 

```python
# Imports
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import pandas as pd
import seaborn as sns
import glob
import os
import time
%matplotlib inline
from pandas.api.types import CategoricalDtype
import warnings
warnings.filterwarnings("ignore")
```

## Step 1: Import data & dimensionality reduction 

```python
# import raw data 
raw = pd.read_csv("../raw_data/MTA_recent_ridership_data.csv")
raw
```

```python
# no prepreocessing required 
raw.describe()
```

```python
y2020 = raw.iloc[217:]
y2020['Date'] = pd.to_datetime(y2020['Date'])
y2020.sort_values(by=['Date'], inplace = True)
y2020
```

```python
# Import taxi data
daily_20 = pd.read_csv("../processed_data/daily_summary_2020.csv").iloc[: , 1:]
daily_20['pickup_date'] = pd.to_datetime(daily_20['pickup_date'])
daily_20
```

## Step 2: Plotting taxi demand and ridership together

```python
# we choose subway 
plt.plot(daily_20.pickup_date, daily_20.trip_count_k, label = 'taxi demand (k)')
plt.plot(y2020['Date'], y2020['Subways: Total Estimated Ridership'] / 10000, label = 'subway ridership (10k)') # NOTE: DIFFERENT SCALE 
plt.xlabel('date')
plt.ylabel('taxi demand (k) / subway ridership (10k)')
plt.title('Daily taxi demand vs subway ridership')
plt.legend()
plt.rcParams["figure.figsize"] = (25,15)
plt.savefig('../plots/taxi_demand_vs_subway_ridership.png')
plt.show()
```
