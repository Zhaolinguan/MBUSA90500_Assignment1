## Q1

Divide each weekday into three periods consisting of a morning rush hour (8:00 - 9:00), lunch hour
(13:00 - 14:00), and evening rush hour (17:00 - 18:00). Each sensor can be associated with a count for each of
these periods. Summing over all sensors we have a period's overall pedestrian count on a particular week day.
Compute the following statistics for each of the three periods in 2022.
- Mean, median, min, max of the period's overall pedestrian count on weekdays

```python
def compute_stats(lst):
    return [np.mean(lst), np.median(lst), min(lst), max(lst)]

def q1():
    count_22 = pd.read_csv("../processed_data/count_22.csv").iloc[:,1:]
    count_22['Date_Time'] = pd.to_datetime(count_22['Date_Time'])
    # omit Saturday and Sunday 
    count_22.drop(count_22.loc[count_22['Day']=='Saturday'].index, inplace=True)
    count_22.drop(count_22.loc[count_22['Day']=='Sunday'].index, inplace=True)
    # define three periods 
    morning_start = 8
    morning_end = 9
    lunch_start = 13
    lunch_end = 14
    evening_start = 17
    evening_end = 18
    # select three time periods 
    mask_morning = (count_22['Time'] >= morning_start) & (count_22['Time'] <= morning_end)
    morning = count_22.loc[mask_morning] 
    mask_lunch = (count_22['Time'] >= lunch_start) & (count_22['Time'] <= lunch_end)
    lunch = count_22.loc[mask_lunch]
    mask_evening = (count_22['Time'] >= evening_start) & (count_22['Time'] <= evening_end)
    evening = count_22.loc[mask_evening]
    # compute stats 
    morning_stats = compute_stats(morning.Hourly_Counts)
    lunch_stats = compute_stats(lunch.Hourly_Counts)
    evening_stats = compute_stats(evening.Hourly_Counts)
    stats_df = pd.DataFrame(list(zip(morning_stats, lunch_stats, evening_stats)), columns =['Morning', 'Lunch', 'Evening'], \
                     index =['mean', 'medium', 'min', 'max'])
    return stats_df

q1()
```

## Q2 

Let the daily overall pedestrian count be the total number of pedestrians counted by sensors on a particular
day. For each of 2021, 2022, generate a scatter plot of maximum temperature (x axis) and daily overall
pedestrian count (y axis)

```python
def daily_count(filename):
    df = pd.read_csv(filename).iloc[:,1:]
    df['Date_Time'] = pd.to_datetime(df['Date_Time'])
    s = df.groupby([df['Date_Time'].dt.date], as_index=True)['Hourly_Counts'].sum()
    new_df = s.to_frame(name="Hourly_Counts")
    new_df.reset_index(inplace=True)
    return new_df
```

```python
def scatter_plot(x, y):
    plt.scatter(x, y)
    plt.show()
```

```python
def q2(): 
    # calculate daily overall pedestrain count
    daily_count_21 = daily_count("../processed_data/count_21.csv")
    daily_count_22 = daily_count("../processed_data/count_22.csv")
    # generate scatter plot
    scatter_plot(temp_21['Maximum temperature (Degree C)'], daily_count_21['Hourly_Counts'])
    scatter_plot(temp_22['Maximum temperature (Degree C)'], daily_count_22['Hourly_Counts'])
    
q2()
```

## Q5

For each of 2021, 2022, generate a histogram showing how busy (mean daily overall pedestrian count y axis)
each day of the week (x axis) is.

```python
def mean_daily_count(df):
    s = df.groupby([df['Date_Time'].dt.date, df['Day']])['Hourly_Counts'].mean()
    new_df = s.to_frame(name="Hourly_Counts")
    new_df.reset_index(inplace=True)
    return new_df
```

```python
SORTER = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
SORTERINDEX = dict(zip(SORTER,range(len(SORTER))))

def groupby_dow(df):
    s = df.groupby(df['Day'], as_index = True)['Hourly_Counts'].mean()
    new_df = s.to_frame(name="Hourly_Counts")
    new_df.reset_index(inplace=True)
    # fill in the void 
    for i in SORTER:
        if i not in new_df['Day'].to_numpy():
            new_df = new_df.append({'Day': i, 'Hourly_Counts': 0}, ignore_index=True)
    new_df['Day'] = new_df['Day'].map(SORTERINDEX)
    new_df.sort_values('Day', inplace=True)
    return new_df
```

```python
def bar_chart(xvalues, data):
    # attribute: https://python-course.eu/numerical-programming/histograms-with-matplotlib.php 
    fig, ax = plt.subplots()
    ax.bar(x=np.arange(len(data)), # The x coordinates of the bars. 
           height=data, # the height(s) of the vars 
           color="green", 
           align="center",
           tick_label=xvalues)
    ax.set_ylabel('mean daily overall pedestrian count')
    ax.set_title('histogram showing how busy...')
    plt.show()
```

```python
def q5(): 
    # calculate daily overall pedestrain count
    mean_daily_count_21 = mean_daily_count(count_21)
    mean_daily_count_22 = mean_daily_count(count_22)
    # sum over day of week 
    dayofweek_21 = groupby_dow(mean_daily_count_21)
    bar_chart(SORTER, dayofweek_21.Hourly_Counts)
    dayofweek_22 = groupby_dow(mean_daily_count_22)
    bar_chart(SORTER, dayofweek_22.Hourly_Counts)
    return dayofweek_21, dayofweek_22
    
d21, d22 = q5()
```

## Q6

For 2022, generate a histogram showing the mean daily overall pedestrian count (y axis) for sensors 1-20 (x
axis).

```python
FIRSTSENSOR = 1
LASTSENSOR = 20

def mean_daily_count_bysensor(df):
    # filter out sensors off interest 
    mask_sensor = (df['Sensor_ID'] >= FIRSTSENSOR) & (df['Sensor_ID'] <= LASTSENSOR)
    df = df.loc[mask_sensor] 
    # take the daily average 
    s = df.groupby([df['Date_Time'].dt.date, df['Sensor_ID']])['Hourly_Counts'].mean()
    new_df = s.to_frame(name="Hourly_Counts")
    new_df.reset_index(inplace=True)
    # groupby by sensors
    s = new_df.groupby(new_df['Sensor_ID'], as_index = True)['Hourly_Counts'].mean()
    new_df = s.to_frame(name="Hourly_Counts")
    new_df.reset_index(inplace=True)
    new_df.sort_values('Sensor_ID', inplace=True)
    return new_df
```

```python
def q5(): 
    # calculate daily overall pedestrain count per sensor 
    mean_21 = mean_daily_count_bysensor(count_21)
    mean_22 = mean_daily_count_bysensor(count_22)
    # plot bar charts 
    bar_chart(mean_21.Sensor_ID, mean_21.Hourly_Counts)
    bar_chart(mean_22.Sensor_ID, mean_22.Hourly_Counts)
    return mean_21, mean_22
    
m1, m2 = q5()
```

## Q7

Same as Q5, but in the calculation only include days where it rained.

```python
def raindays(count_df, rain_df):
    merged = count_df.merge(rain_df, left_on='Date_Time', right_on='Date')
    result = merged.loc[merged['Quality'] == 'Y']
    return result
```

```python
def q7(): 
    # identify rain days
    countmergerain_21 = raindays(count_21, rain_21)
    countmergerain_22 = raindays(count_22, rain_22)
    # calculate daily overall pedestrain count
    mean_21 = mean_daily_count(countmergerain_21)
    mean_22 = mean_daily_count(countmergerain_22)
    # sum over day of week 
    dayofweek_21 = groupby_dow(mean_21)
    bar_chart(SORTER, dayofweek_21.Hourly_Counts)
    dayofweek_22 = groupby_dow(mean_22)
    bar_chart(SORTER, dayofweek_22.Hourly_Counts)
    return dayofweek_21, dayofweek_22
    
d1, d2 = q7()
```

```python
def maxtempless20(df, temp_df):
    merged = df.merge(temp_df, left_on='Date_Time', right_on='Date')
    result = merged.loc[merged['Maximum temperature (Degree C)'] < 20]
    return result
```

```python
def q8(): 
    # identify rain days
    countmergerain_21 = raindays(count_21, rain_21)
    countmergerain_22 = raindays(count_22, rain_22)
    # out of rained days, identify days with temperature less than 20 
    merged_21 = maxtempless20(countmergerain_21, temp_21)
    merged_22 = maxtempless20(countmergerain_22, temp_22)
    # calculate daily overall pedestrain count
    mean_21 = mean_daily_count(merged_21)
    mean_22 = mean_daily_count(merged_22)
    # sum over day of week 
    dayofweek_21 = groupby_dow(mean_21)
    bar_chart(SORTER, dayofweek_21.Hourly_Counts)
    dayofweek_22 = groupby_dow(mean_22)
    bar_chart(SORTER, dayofweek_22.Hourly_Counts)
    return dayofweek_21, dayofweek_22
    
d1, d2 = q8()
```
