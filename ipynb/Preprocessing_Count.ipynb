# Preprocess count data 


```python
# Imports
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import pandas as pd
import seaborn as sns
import glob
import os
import time
%matplotlib inline
from pandas.api.types import CategoricalDtype
import warnings
warnings.filterwarnings("ignore")
```

## Step 1: Import data & feature selection

```python
# import raw data 
count_raw = pd.read_csv("../raw_data/count2021-2022.csv")
count_raw['Date_Time'] = pd.to_datetime(count_raw['Date_Time'])
count_raw
```

```python
count_raw.describe()
```

```python
start_date_2021 = '2021-01-01'
end_date_2021 = '2021-12-31'
start_date_2022 = '2022-01-01'
end_date_2022 = '2022-05-31'
```

```python
# find all 2021 data 
mask_21 = (count_raw['Date_Time'] >= start_date_2021) & (count_raw['Date_Time'] <= end_date_2021)
count_21 = count_raw.loc[mask_21] 

# find all 2022 data 
mask_22 = (count_raw['Date_Time'] >= start_date_2022) & (count_raw['Date_Time'] <= end_date_2022)
count_22 = count_raw.loc[mask_22] 
```

## Step 2: Save on desk

```python
count_21.to_csv('../processed_data/count_21.csv',header=True)
count_22.to_csv('../processed_data/count_22.csv',header=True)
```
